```python
# Membuat data frame dan menampilkan data
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Semester3/Peminatan/Data_Mining/Dataset/bakery_sales_revised.csv')
df
```

```python
# Mengubah data ke dalam format one-hot encoding
# Membuat tabel pivot
df_pivot= df.pivot_table(index='Transaction', columns='Item', aggfunc='size', fill_value=0)
df_pivot
```

```python
# Mengubah ke bentuk binary (jika item dibeli labelnya 1, jika tidak labelnya 0)
df_binary = df_pivot.applymap(lambda x: 1 if x > 0 else 0)
df_binary
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
# Mencari itemsets yang sering muncul menggunakan algoritma apriori minimal 0,09
frequent_itemsets = apriori(df_binary, min_support=0.09, use_colnames=True)

# Menarapkan aturan asosiasi dan menampilkan hasil berdasarkan lift dengan minimal thresholdnya 0,1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.1)
```
