```python
import pandas as pd
import numpy as np

# Baca file CSV
df = pd.read_csv('/500_Person_Gender_Height_Weight_Index.csv')  # Sesuaikan dengan path file tempat naro file csvnya
print(df)
```

```python
# Menghitung statistik berdasarkan gender
# hitung total data berdasarkan gender
count_by_gender = df['Gender'].value_counts()
print("Jumlah data berdasarkan gender:")
print(count_by_gender)

# Menghitung statistik tinggi dan berat berdasarkan gender
stats = df.groupby('Gender').agg({
    'Height': ['mean', 'min', 'max'],  # Statistik untuk tinggi (rata-rata, min, max)
    'Weight': ['mean', 'min', 'max']   # Statistik untuk berat badan (rata-rata, min, max)
})

print("\nStatistik tinggi dan berat berdasarkan gender:")
print(stats)
```

```python
df['Gender'] = df['Gender'].replace({'Female': 0, 'Male': 1}).astype(int)
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Index', hue='Gender', data=df)

# Menambahkan judul dan label
plt.title('Visualisasi Index BMI Base on Gender')
plt.xlabel('Indeks BMI')
plt.ylabel('Jumlah')

# Menambahkan penjelasan kategori BMI
plt.xticks(ticks=[0, 1, 2, 3, 4, 5],
           labels=['Extremely\nWeak(0)', 'Weak\n(1)', 'Normal\n(2)', 'Overweight\n(3)', 'Obesity\n(4)', 'Extreme\nObesity\n(5)'])

# Menampilkan grafik
plt.show()
```
